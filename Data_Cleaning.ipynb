## Data Cleaning

```python
#Removing unnecessary columns 
df = df[[#Unnamed: 0,  'id', 'budget', 'revenue','tagline', 'keywords','homepage','overview',
         'imdb_id', 'popularity',
       'original_title', 'cast',  'director', 
        'runtime', 'genres', 'production_companies', 'release_date',
       'vote_count', 'vote_average', 'release_year', 'budget_adj',
       'revenue_adj', 'profit', 'popularity_level']].copy()
df.head()
```

```python
#Changing release_date dtypes to datetime and profit, revenue and budget to numeric
df['release_date'] = pd.to_datetime(df['release_date'])
df['revenue_adj'] = pd.to_numeric(df['revenue_adj'])
df['budget_adj'] = pd.to_numeric(df['budget_adj'])
df['profit'] = pd.to_numeric(df['profit'])
df.dtypes
```

## Data Manipulation

Cleaning columns with multiple value

- Converting columns with multiple values, and set it to 1 value for genres, production_companies and casts
- The first value of the columns will be considered as the 'main' values that will be used for further analysis

```python
#For columns genres, cast and production_companies, only keeps the first value
#Regex Pattern for genres : \w*[^\|]
#Regex Pattern for cast and production_companies : .*?\|

#genres
df['genres'] = df['genres'].str.extract(r'(\w*[^\|])', expand=False)

#company
df['production_companies'] = df['production_companies'].str.extract(r'(.*?\|)', expand=False)
df['production_companies'] = df['production_companies'].str.replace('|','')

#cast
df['cast'] = df['cast'].str.extract(r'(.*?\|)', expand=False)
df['cast'] = df['cast'].str.replace('|','')

df.head()
```

Add category code for genres, cast and production_companies, director and popularity level. 

```python
#production company
df['production_companies'] = df['production_companies'].astype('category') 
df['company_code'] = df['production_companies'].cat.codes
```

```python
#cast
df['cast'] = df['cast'].astype('category') 
df['cast_code'] = df['cast'].cat.codes
```

```python
#genres
df['genres'] = df['genres'].astype('category') 
df['genres_code'] = df['genres'].cat.codes
```

```python
#director
df['director'] = df['director'].astype('category') 
df['director_code'] = df['director'].cat.codes
```

```python
#popularity level, customize num cat because it shows a degree of value

#creating dict for popularity level dictionary
pop_lvl_name = df.popularity_level.unique().tolist()
pop_lvl_num = (['4','3','2','1'])
pop_level_ref = dict(zip(pop_lvl_name,pop_lvl_num))

#assign dictionary to new column
df['popularity_code'] = df['popularity_level'].map(pop_level_ref)
```
